# __Research Question #1:__
### How was Derek Jeter's base-stealing efficiency over the course of his career and at what age did he have a career high in stolen bases?

```python
#Calculate characteristics of Derek Jeter's stolen base and caught stealing

total_attempts = jeter_stats_df["SB"].sum() + jeter_stats_df["CS"].sum()
stolen_bases_total = jeter_stats_df["SB"].sum()
caught_stealing_total = jeter_stats_df["CS"].sum()
most_bases_stealed = jeter_stats_df["SB"].max()
most_caught_stealing = jeter_stats_df["CS"].max()

#Efficiency calculated
stolen_base_percentage = round((stolen_bases_total/total_attempts) *100) # https://www.freecodecamp.org/news/how-to-round-numbers-up-or-down-in-python/
caught_stealing_percentage = round((caught_stealing_total/total_attempts) *100)

# sb = stolen bases
#locate row in which Derek Jeter had his most stolen bases
#get career high in stolen bases and the age from the row
career_high_sb_row = jeter_stats_df.loc[jeter_stats_df['SB'].idxmax()]
career_high_sb = career_high_sb_row['SB']
career_high_sb_age = career_high_sb_row['Age']

print(
    f"Derek Jeter attempted to steal a base {total_attempts} times in his career.\n" #\n == new line
    f"He stole a total of {stolen_bases_total} bases in his career, making him safe {stolen_base_percentage}% of the time.\n"
    f"He was caught stealing {caught_stealing_total} times in his career, which was {caught_stealing_percentage}% of his attempts.\n"
    f"Derek's career high was {career_high_sb} stolen bases in his age {career_high_sb_age} season."
)
```

```python
# Scatter plot for Derek Jeter's stolen bases over his career
plt.figure(figsize=(10, 6))
plt.scatter(jeter_stats_df['Year'], jeter_stats_df['SB'], color='navy', label='Stolen Bases', marker='o')

# Add labels, title, and custom ticks
plt.xlabel('Year')
plt.ylabel('Stolen Bases')
plt.title("Derek Jeter's Stolen Bases Over His Career", color= 'navy')
plt.xticks(jeter_stats_df['Year'], rotation=45)  # Rotate the year labels for better readability
#https://www.geeksforgeeks.org/matplotlib-pyplot-xticks-in-python/
plt.yticks(range(0, jeter_stats_df['SB'].max()+1, 2))
plt.legend()
plt.grid(True)

# Display the plot
plt.tight_layout()
plt.show()


# Now a different scatter plot showing his caught stealing per season

plt.figure(figsize=(10, 6))
plt.scatter(jeter_stats_df['Year'], jeter_stats_df['CS'], color='orange', label='Caught Stealing', marker='o')

# Add labels, title, and custom ticks
plt.xlabel('Year')
plt.ylabel('Caught Stealing')
plt.title("How many times Derek Jeter was Caught Stealing in his Career per season", color= 'orange')
plt.xticks(jeter_stats_df['Year'], rotation=45)  # Rotate the year labels for better readability
plt.yticks(range(0, jeter_stats_df['CS'].max()+1, 1))
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()


# Grouped Bar Chart differentiating how many stolen bases vs how many times he was caught stealing per year
# https://www.geeksforgeeks.org/bar-plot-in-matplotlib/
# https://www.geeksforgeeks.org/plotting-multiple-bar-charts-using-matplotlib-in-python/ 

width = 0.375  # Width of the bars
x = range(len(jeter_stats_df['Year']))  # Label locations

fig, ax = plt.subplots(figsize=(12, 6))
bars1 = ax.bar([pos - width/2 for pos in x], jeter_stats_df['SB'], width, label='Stolen Bases', color='navy') # shifts position to left
bars2 = ax.bar([pos + width/2 for pos in x], jeter_stats_df['CS'], width, label='Caught Stealing', color='orange') # shifts position to right


# Add labels, title, and custom ticks
ax.set_xlabel('Year')
ax.set_ylabel('Count')
ax.set_title("Derek Jeter's Stolen Bases vs Caught Stealing by Year")
ax.set_xticks(x)
ax.set_xticklabels(jeter_stats_df['Year'], rotation=45)
ax.legend()

plt.tight_layout() # used https://matplotlib.org/stable/users/explain/axes/tight_layout_guide.html# to determine how to fit 
plt.show()
```

# __Research Question #3:__
### What is the breakdown of the outcomes of all of Derek Jeter's at-bats throughout his career? Specifically, how often did he walk, strike out, hit a single, double, triple, etc.?

```python
# Pie chart to breakdown what Derek Jeter did in his at bats
# Used lab slides and https://www.w3schools.com/python/matplotlib_pie_charts.asp to help configure chart

# Calculate the total number of at-bats in Jeter's career
total_ab = jeter_stats_df['AB'].sum()

# Calculate the total number of hits and each specific type of hit
total_hits = jeter_stats_df['H'].sum()  # Total hits (includes singles, doubles, triples, and home runs)
total_doubles = jeter_stats_df['2B'].sum()  # Total doubles
total_triples = jeter_stats_df['3B'].sum()  # Total triples
total_home_runs = jeter_stats_df['HR'].sum()  # Total home runs

#No category for just singles so we must calculate ourselves
total_singles = total_hits - (total_doubles + total_triples + total_home_runs)

# Calculate the total number of walks, strikeouts, and grounded into double plays
total_walks = jeter_stats_df['BB'].sum()
total_strikeouts = jeter_stats_df['SO'].sum()
total_doubleplays = jeter_stats_df['GDP'].sum()

#There are other that are not on this list but also count as at bats
#These other at bat outcomes could be flyouts, popouts, lineout, groundouts, etc., so we will just call them other outcomes
total_other_possibilities = total_ab - (total_hits + total_walks +total_strikeouts+total_doubleplays)

# Calculate the percentage breakdown of each outcome relative to the total at-bats
singles_percent = (total_singles/total_ab)*100
doubles_percent = (total_doubles/total_ab)*100
triples_percent = (total_triples/total_ab)*100
home_runs_percent = (total_home_runs/total_ab)*100
walks_percent = (total_walks/total_ab)*100
strikeout_percent = (total_strikeouts/total_ab)*100
double_play_percent = (total_doubleplays/total_ab)*100
other_possibilities_percent = (total_other_possibilities/total_ab)*100

# Labels for each slice of the pie chart, indicating each category of at-bat outcome
labels = ["Singles", "Doubles", "Triples", "Home Runs", "Walks", "Strikeouts", "Grounded into Double Plays","Other Outcomes"]
numbers = [singles_percent, 
           doubles_percent, 
           triples_percent, 
           home_runs_percent, 
           walks_percent, 
           strikeout_percent, 
           double_play_percent,
           other_possibilities_percent]

# Explode the "Other Outcomes" section slightly to make it stand out
explode = (0, 0, 0, 0, 0, 0, 0, 0.1) #Explode the Other outcomes region to show that the outcome is not specified  

# Add hatches to show the total hits
hatches = ["/", "/", "/", "/", "", "", "", ""]  

# Add colors for each slice of the pie chart
mycolors = ["darkturquoise", "slateblue", "lime", "crimson", "pink", "violet", "lightsalmon", "purple"]  

fig, ax = plt.subplots(figsize=(10, 8))  # Larger figure size for clarity
ax.pie(numbers, explode=explode, labels=labels, hatch=hatches, autopct="%1.1f%%", colors=mycolors,  
    shadow=False, 
    startangle=90
)
ax.axis("equal")  # Ensures the pie chart is a circle
plt.title("Derek Jeter's At-Bat Breakdown", fontsize=16)  

#Used https://how2matplotlib.com/matplotlib-pie-chart-legend.html# 
#And https://saturncloud.io/blog/legend-outside-the-plot-in-python-matplotlib/#:~:text=To%20move%20the%20legend%20outside,parameters%20of%20the%20legend%20function.&text=The%20bbox_to_anchor%20parameter%20specifies%20the,legend's%20anchor%20point%20should%20be.
plt.legend(bbox_to_anchor=(1.05, 1))  # added legend to follow colors more accurately

plt.show()
```

#### The pie chart illustrates the distribution of outcomes for all of Derek Jeter's at-bats throughout his career:

* Hits: Out of his at bats 31% of them were hits as illustrated by the hatched portion of the chart. This 31% means that his career batting average was .310 which can be confirmed by the website Baseball Refernce. https://www.baseball-reference.com/players/j/jeterde01.shtml
    * Singles: The most common outcome, accounting for the majority of his hits.
    * Doubles: Represent a smaller but significant portion of his at-bats, showcasing his ability to hit for extra bases.
    * Triples: A rare occurrence, reflecting his speed and situational hitting.
    * Home Runs: A modest percentage, consistent with his role as a contact hitter rather than a power hitter.
* Walks: Highlight his plate discipline and ability to get on base without putting the ball in play.
* Strikeouts: A relatively small percentage in relation to total at bats, demonstrating his contact-oriented approach.
* Grounded into Double plays: This accounts for a small percentage of his at-bats, comparable to his home run percentage. This suggests that he didn't often hit ground balls in double-play situations (fewer than two outs with at least one runner on base, making force plays possible). It also reflects his speed, as he likely beat out many potential double plays, turning them into fielder's choices or singles.
* Other Outcomes: Includes outs made via groundouts, flyouts, and other means not specified, making up the remainder of his at-bats.

This breakdown provides insight into Jeter's offensive profile, emphasizing his proficiency at making contact and reaching base.
