```python
import time
import random
```

```python
class Player:
    def __init__(self, name):
        self.name = name
        self.money = 100  # Starting money
        self.inventory = []
        self.animal_relationship = 0
        self.coziness_level = 50  # Coziness level starts high
        self.animals = {
            "chicken": 2,
            "pig": 1,
            "horse": 1,
            "cow": 2,
            "goat": 0
        }


    def decrease_coziness(self):
        self.coziness_level -= 1
        if self.coziness_level < 0:
            self.coziness_level = 0
        print(f"Coziness level is now {self.coziness_level}")

    def increase_coziness(self, amount):
        self.coziness_level += amount
        if self.coziness_level > 100:  # Max coziness cap
            self.coziness_level = 100
        print(f"You gained coziness! Coziness level is now {self.coziness_level}")

    def add_to_inventory(self, item):
        self.inventory.append(item)

    def show_inventory(self):
        if not self.inventory:
            print("Your inventory is empty.")
        else:
            print("Inventory:")
            for item in self.inventory:
                print(f"- {item}")

    def add_money(self, amount):
        self.money += amount
        print(f"You earned ${amount}. Total money: ${self.money}")

    def spend_money(self, amount):
        if self.money >= amount:
            self.money -= amount
            print(f"You spent ${amount}. Remaining money: ${self.money}")
            return True
        else:
            print("Not enough money.")
            return False

    def add_animal(self, animal_name):
        if animal_name in self.animals:
            self.animals[animal_name] += 1
        else:
            self.animals[animal_name] = 1
        print(f"Added a {animal_name}. Total {animal_name}s: {self.animals[animal_name]}")

    def show_animals(self):
        print("Animals on the farm:")
        for animal, count in self.animals.items():
            print(f"- {animal.capitalize()}: {count}")

    def animal_rship_increase(self):
          self.animal_relationship += 1
          print(f"Your relation with the animals is now {self.animal_relationship}")
          if self.animal_relationship == 9:
            special_items = ["Redbull", "Beachball", "Rollerskate", "Mona Lisa", "Golden Egg"]
            special_item_found = random.sample(special_items, 1)
            for special_items in special_item_found:
              self.inventory.append(special_item_found)
              print(f"Your relationship with the animals have granted you a reward! A {special_item_found}!")
              self.coziness_level += 5
              print(f"The {special_item_found} granted you 5 coziness!")



```

```python
class Tool:
    def __init__(self, name, efficiency, cost, level_required):
        self.name = name
        self.efficiency = efficiency
        self.cost = cost
        self.level_required = level_required

    def use_tool(self):
        print(f"Using {self.name} tool.")
```

```python
class Plant:
    PLANT_PRICES = {
        "wheat": 5,
        "carrots": 8,
        "tomatoes": 10,
        "potatoes": 7,
        "pumpkins": 15,
        "apples": 12,
        "peaches": 14,
        "strawberries": 10,
        "blueberries": 10,
        "flowers": 6
    }

    def __init__(self, name):
        self.name = name
        self.health = 100
        self.nutrients = 50
        self.water = 50

    def decrease_health(self):
        self.health -= 2  # Decrease health each day
        if self.health < 0:
            self.health = 0
        print(f"{self.name}'s health is now {self.health}.")

    def check_health(self):
        print(f"{self.name} - Health: {self.health}, Water: {self.water}, Nutrients: {self.nutrients}")
```

```python
class Animal:
    ANIMAL_PRICES = {
        "chicken": 10,
        "pig": 50,
        "horse": 100,
        "cow": 75,
        "goat": 40
    }

    def __init__(self, species):
        self.species = species
```

```python
class Garden:
    def __init__(self):
        self.water_level = 50
        self.fertilizer_level = 50
        self.pest_level = 0
        self.plants = {  # Dictionary to keep track of each plant type and count
            "wheat": 0,
            "carrots": 0,
            "tomatoes": 0,
            "potatoes": 0,
            "pumpkins": 0,
            "apples": 0,
            "peaches": 0,
            "strawberries": 0,
            "blueberries": 0,
            "flower": 0
        }

    def add_plant(self, plant_name):
        if plant_name in self.plants:
            self.plants[plant_name] += 1
        else:
            self.plants[plant_name] = 1
        print(f"Added a {plant_name}. Total {plant_name}s: {self.plants[plant_name]}")

    def show_plants(self):
        print("Plants in the garden:")
        for plant, count in self.plants.items():
            print(f"- {plant.capitalize()}: {count}")

    def water_plant(self):
        self.water_level = 50
        self.water_level += 10
        print(f"Watered Garden. Water level is now {self.water_level}.")

    def fertilize(self):
        self.fertilizer_level = 50
        self.fertilizer_level += 10
        print(f"Fertilized Plants. Nutrient level is now {self.fertilizer_level}.")

    def decrease_garden_status(self):
        self.water_level -= 1                            # Decrease each day
        self.fertilizer_level -= 1
        if self.water_level < 0:
            self.water_level = 0
        if self.fertilizer_level < 0:
            self.fertilizer_level = 0
        print(f"Garden Status - Water Level: {self.water_level}, Fertilizer Level: {self.fertilizer_level}")

    def check_garden_status(self):
        print("Garden Status:")
        print(f"Water Level: {self.water_level}")
        print(f"Fertilizer Level: {self.fertilizer_level}")

```

```python
class Shop:
    TOOL_PRICES = {
        "pickaxe": 20,
        "hoe": 15,
        "shovel": 10
    }

    def __init__(self):
        self.plant_prices = Plant.PLANT_PRICES
        self.animal_prices = Animal.ANIMAL_PRICES

    def sell_item(self, player, item, amount=1):
        if isinstance(item, Plant):
            price = self.plant_prices.get(item.name, 0) * amount
            player.add_money(price)
            print(f"Sold {amount} {item.name}(s) for ${price}.")
        elif isinstance(item, Animal):
            price = self.animal_prices.get(item.species, 0) * amount
            player.add_money(price)
            print(f"Sold {amount} {item.species}(s) for ${price}.")
        else:
            print("Cannot sell this item.")

    def buy_tool(self, player, tool_name):
        price = self.TOOL_PRICES.get(tool_name.lower())
        if price:
            if player.spend_money(price):
                tool = Tool(tool_name, efficiency=10, cost=price, level_required=1)
                player.add_to_inventory(tool_name)
                print(f"Bought {tool_name} for ${price}.")
            else:
                print(f"You do not have enough money to buy {tool_name}.")
        else:
            print("Tool not available in the shop.")

    def buy_animal(self, player, animal_name):
        random_price = random.randint(Animal.ANIMAL_PRICES[animal_name] - 10, Animal.ANIMAL_PRICES[animal_name] + 10)
        if player.spend_money(random_price):
            player.add_animal(animal_name)
            print(f"Bought a {animal_name} for ${random_price}.")
        else:
            print(f"You do not have enough money to buy {animal_name}.")


def explore_forest(player, garden):
    # Finding random animals
    forest_animals = ["chicken", "pig", "cow", "horse", "goat"]
    num_animals_found = random.randint(1, 5)
    found_animals = random.choices(forest_animals, k=num_animals_found)

    print(f"In the forest, you found {num_animals_found} animal(s):")
    for animal in found_animals:
        player.add_animal(animal)

    # Finding random plants in forest
    forest_plants = ["wheat", "carrots", "tomatoes", "potatoes", "pumpkins", "apples", "peaches", "strawberries", "blueberries", "flowers"]
    num_plants_found = random.randint(1, 6)
    found_plants = random.choices(forest_plants, k=num_plants_found)

    print(f"You also found {num_plants_found} plant(s):")
    for plant in found_plants:
        garden.add_plant(plant)

    # Finding cozy items in forest
    cozy_items = ["Sweater", "Fuzzy Keychain", "Blanket", "Authority Cap", "Fuzzy Socks"]
    cozyitem_found = random.sample(cozy_items, 1)
    for cozy_item in cozyitem_found:
      player.add_to_inventory(cozyitem_found)
      player.increase_coziness(7)
      print(f"{cozyitem_found} was found!")

def sell_at_market(player, garden, shop):
    print("Selling at market:")

    # Option to sell animals
    print("Choose an animal to sell:")
    player.show_animals()
    animal_choice = input(f"Enter the animal you want to sell (or type 'none' to skip): ").lower()
    if animal_choice != 'none' and animal_choice in player.animals and player.animals[animal_choice] > 0:
        amount = int(input(f"How many {animal_choice}s do you want to sell? "))
        if amount <= player.animals[animal_choice]:
            shop.sell_item(player, Animal(animal_choice), amount)
        else:
            print(f"You don't have that many {animal_choice}s.")

    # Option to sell plants
    print("Choose a plant to sell:")
    garden.show_plants()
    plant_choice = input(f"Enter the plant you want to sell (or type 'none' to skip): ").lower()
    if plant_choice != 'none' and plant_choice in garden.plants and garden.plants[plant_choice] > 0:
        amount = int(input(f"How many {plant_choice}s do you want to sell? "))
        if amount <= garden.plants[plant_choice]:
            plant = Plant(plant_choice)
            shop.sell_item(player, plant, amount)
        else:
            print(f"You don't have that many {plant_choice}s.")

    # Option to buy animals at the market with random prices
    print("Animals available for purchase at the market:")
    for animal in Animal.ANIMAL_PRICES.keys():
        print(f"- {animal.capitalize()}: Estimated price ${Animal.ANIMAL_PRICES[animal]}")

    animal_to_buy = input("Enter the animal you want to buy (or type 'none' to skip): ").lower()
    if animal_to_buy != 'none' and animal_to_buy in Animal.ANIMAL_PRICES:
        shop.buy_animal(player, animal_to_buy)
```

```python

def main():
    player_name = input("Enter your name: ")
    player = Player(player_name)
    garden = Garden()
    shop = Shop()
    plant = Plant("Garden")
    animal = Animal("test")


    print(f"Welcome to the farming Adevnture, {player.name}!")
    print("In this relaxing farming simulation game, you get to manage your own farm and explore a variety of exciting locations. Wander through your garden, where you can plant and nurture crops, or venture into the forest to gather resources. Visit the town to sell your items and interact with the locals. Tend to your animals, care for your plants, and watch your farm grow as you work to build your dream farm. Whether you're harvesting crops, tending to animals, or making a trip to the market, there's always something new to discover!")

    while True:
        print("Choose an activity:")
        print("1. Tend Garden")
        print("2. Explore Town")
        print("3. Tend Animals")
        print("4. Explore Woods")
        print("5. Quit")

        choice = input("Enter the number of your choice: ")

        if choice == "1":
            print("\nTend Garden:")
            print("1. Water plants")
            print("2. Fertilize plants")
            print("3. Check garden status")
            print("4. Check Inventory")


            garden_choice = input("Choose an option: ")
            if garden_choice == "1":
                garden.water_plant()
            elif garden_choice == "2":
                garden.fertilize()
            elif garden_choice == "3":
                garden.check_garden_status()
            elif garden_choice == "4":
              player.show_inventory()

        elif choice == "2":
            print("Explore Town:")
            print("1. Go to shop")
            print("2. Sell at market")

            town_choice = input("Choose an option: ")

            if town_choice == "1":
                print("Welcome to the shop! Available tools:")
                for tool, price in shop.TOOL_PRICES.items():
                    print(f"{tool.capitalize()} - ${price}")
                tool_choice = input("Enter the tool you want to buy: ")
                shop.buy_tool(player, tool_choice)
            elif town_choice == "2":
                sell_at_market(player, garden, shop)

        elif choice == "3":
            print("Animals on the farm:")
            player.show_animals()
            player.animal_rship_increase()

        elif choice == "4":
            explore_forest(player, garden)

        elif choice == "5":
            print("Thanks for playing!")
            break

        else:
            print("Invalid option. Try again.")

        # Simulate a day passing
        player.decrease_coziness()
        garden.decrease_garden_status()

main()
```
